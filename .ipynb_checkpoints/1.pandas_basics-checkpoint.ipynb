```python
import pandas as pd

df = pd.read_csv("data.csv")
df
```

# Using dictionary as data source

```python
data_source = {
    "name" : ['ram', 'shyam', 'chaman', 'virapan', 'thirpu'],
    "age" : [34,45,56,67,78],
    "salary" : [5000,6000,2000,4000,1000]
}
```

```python
df = pd.DataFrame(data_source)
df
```

# Properties : Shape

```python
# Shape gives (rows, column) as Tuple
df.shape

# using tuple values
rows, columns = df.shape
print(rows,columns)
```

# Properties : Head()

```python
# Head prints initial few rows (just to give some idea about the data)
df.head()
```

```python
# Passing parameter to head()
df.head(3)

# will print initial 3 rows
```

# Properties : tail()

```python
# print few rows from end
df.tail()

# we can also pass parameter to tail function
df.tail(1)

# will print only last one row
```

# Using Slicing [m:n]

```python
df[2:4]
```

# Some Important functions

# max(), min(), mean(), std()

```python
df.energy.max()
# or 
df['energy'].max()

df.energy.min()
df.energy.std()
```

# Describe() : Print statistics about Integer data

```python
df.describe()
```

# Sql like Querying

```python
df[df.energy > 0]
```

```python
# Printing row with maximium energy
df[df.energy == df.energy.max()]
```

```python
# Printing specific column 
df.columns
df[['key', 'energy', 'liveness']][df.energy == df.energy.max()]
```
